### 3. Но и это ещё не всё: биграммы скорее всего тоже далеко не всегда работают. Основная часть задания — в том, как можно их улучшить:
### - предложите метод обучения перестановки символов в этом задании, основанный на MCMC-сэмплировании, но по-прежнему работающий на основе статистики биграмм;

```python
import string
import math
import random

# функция для создания словаря шифрования
def create_cipher_dict(cipher, text):
    cipher_dict = {}
    alphabet_list = sorted(list(set(text)))
    for i in range(len(alphabet_list)):
        cipher_dict[alphabet_list[i]] = cipher[i]
    return cipher_dict

# функция для шифрования текста
def apply_cipher_on_text(text, cipher):
    cipher_dict = create_cipher_dict(cipher, text) 
    text = list(text)
    newtext = ""
    for elem in text:
            newtext += cipher_dict[elem]
    return newtext
```

```python
string_to_encode = string_example
print("String_to_encode:", string_to_encode)
print("\n")

encryption_key = ''.join(random.sample(list(set(string_to_encode)), 31))
print("Encryption_key:", encryption_key)
print("\n")

string_to_decode = apply_cipher_on_text(string_to_encode, encryption_key)
print("String_to_decode:", string_to_decode)
```

```python
# считаем частоту биграм после применения шифра
def score_params_on_cipher(text):
    return Counter(text[idx : idx + 2] for idx in range(len(text) - 1))

# рассчитываем score шифра
def get_cipher_score(text, cipher, scoring_params):
    cipher_dict = create_cipher_dict(cipher, text)
    decrypted_text = apply_cipher_on_text(text,cipher)
    scored_f = score_params_on_cipher(decrypted_text)
    cipher_score = 0
    for k, v in scored_f.items():
        if k in scoring_params:
            cipher_score += v * math.log(scoring_params[k])
    return cipher_score
```

```python
# случайно геерируем шифр
def generate_cipher(cipher):
    pos1 = random.randint(0, len(list(cipher)) - 1)
    pos2 = random.randint(0, len(list(cipher)) - 1)
    if pos1 == pos2:
        return generate_cipher(cipher)
    else:
        cipher = list(cipher)
        cipher[pos1], cipher[pos2] = cipher[pos2], cipher[pos1]
        return "".join(cipher)

def random_coin(p):
    unif = random.uniform(0, 1)
    if unif >= p:
        return False
    else:
        return True
```

```python
def MCMC_decrypt(n_iter, cipher_text, scoring_params):
#     current_cipher = "ЫПМСЮБГДШ ЖЙЬЧРЗОЯЛВИФЕТАЦКЩУХН"
    current_cipher = ''.join(random.sample(list(set(cipher_text)), len(list(set(cipher_text)))))
    state_keeper = set()
    best_state = ''
    score = 0
    for i in range(n_iter):
        state_keeper.add(current_cipher)
        proposed_cipher = generate_cipher(current_cipher)
        while proposed_cipher in state_keeper:
            proposed_cipher = generate_cipher(current_cipher)
        score_current_cipher = get_cipher_score(cipher_text,current_cipher,scoring_params)
        score_proposed_cipher = get_cipher_score(cipher_text,proposed_cipher,scoring_params)
        acceptance_probability = min(1, math.exp(score_proposed_cipher - score_current_cipher))
        if score_current_cipher > score:
            best_state = current_cipher
        if random_coin(acceptance_probability):
            current_cipher = proposed_cipher
        if i % 1000 == 0:
            print("iter", i, ":",apply_cipher_on_text(cipher_text,current_cipher)[0:99])
    return state_keeper,best_state
```
