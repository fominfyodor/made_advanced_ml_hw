## 3. Качество рейтинг-системы оценивается качеством предсказаний результатов турниров. Но сами повопросные результаты наши модели предсказывать вряд ли смогут, ведь неизвестно, насколько сложными окажутся вопросы в будущих турнирах; да и не нужны эти предсказания сами по себе. Поэтому предложите способ предсказать результаты нового турнира с известными составами, но неизвестными вопросами, в виде ранжирования команд; в качестве метрики качества на тестовом наборе давайте считать ранговые корреляции Спирмена и Кендалла (их можно взять в пакете scipy) между реальным ранжированием в результатах турнира и предсказанным моделью, усреднённые по тестовому множеству турниров.


```python
# мерджим силу игрока и номер игрока
df_results_all_team_players = df_results_all_team_players.merge(df_player_s, left_on='id_player', right_on='id_player')
df_results_all_team_players
```

```python
# меняем порядок у силы игрока
df_results_all_team_players['power_player'] = -1 * df_results_all_team_players['power_player']
```

```python
# # отбираем данные для тестового датасета
# df_tour_team_pos_test = df_tour_team_pos.loc[(df_tour_team_pos['dateStart'] >= '2020-01-01')]
# df_tour_team_pos_test
```

```python
# отбираем данные для тестового датасета
df_results_all_team_players_test = df_results_all_team_players.loc[(df_results_all_team_players['dateStart'] >= '2020-01-01')]
df_results_all_team_players_test
```

```python
# предполагаем, что рейтинг команды на турнире = силе самого сильного игрока 
df_groupby_1 = df_results_all_team_players_test.groupby(['id_tournament', 'id_team', 'position'])[['power_player']].max().reset_index()
df_groupby_1
```

```python
df_pos = df_groupby_1.groupby('id_tournament')['position'].apply(list).to_dict()
df_power_player = df_groupby_1.groupby('id_tournament')['power_player'].apply(list).to_dict()
```

```python
# считаем коррелицию Спирмена 
from scipy import stats
list_spearmanr = []
for i in df_pos:
    list_spearmanr.append(stats.spearmanr(df_pos[i], df_power_player[i])[0])
    
cleaned_list_spearmanr = [x for x in list_spearmanr if str(x) != 'nan']
print('Корреляция Спирмена =', sum(cleaned_list_spearmanr) / len(cleaned_list_spearmanr))
```

```python
# считаем коррелицию Кендалла 
from scipy import stats
list_kendall = []
for i in df_pos:
    list_kendall.append(stats.kendalltau(df_pos[i], df_power_player[i])[0])
    
cleaned_list_kendall = [x for x in list_kendall if str(x) != 'nan']
print('Корреляция Кендалла =', sum(cleaned_list_kendall) / len(cleaned_list_kendall))
```
